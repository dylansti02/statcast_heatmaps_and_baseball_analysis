```python
batted_data = pd.read_csv('batted_balls_2025.csv.gz', compression='gzip')
```

```python
def categorize_batted_ball_direction(row):
    if pd.isnull(row['hc_x']) or pd.isnull(row['stand']):
        return np.nan
    if row['stand'] == 'L':
        if row['hc_x'] > 190:  # pull
            return 'Pull'
        elif row['hc_x'] < 120:  # oppo
            return 'Oppo'
        else:
            return 'Center'
    else:  # Right-handed batter
        if row['hc_x'] < 120:  # pull
            return 'Pull'
        elif row['hc_x'] > 190:  # oppo
            return 'Oppo'
        else:
            return 'Center'
```

```python
pb.cache.disable()
```

```python
batted_data['direction']=batted_data.apply(categorize_batted_ball_direction, axis=1)
```

```python

filtered = batted_data.dropna(subset=['direction', 'bb_type'])


grouped = filtered.groupby(['batter', 'direction', 'bb_type']).size().reset_index(name='count')


pivoted = grouped.pivot_table(
    index='batter',
    columns=['direction', 'bb_type'],
    values='count',
    fill_value=0
)


pivoted.columns = [f"{direction}_{bb_type}" for direction, bb_type in pivoted.columns]


pivoted = pivoted.reset_index()

```

```python
pivoted
```

```python
proportions = pivoted.copy()
proportions.iloc[:, 1:] = proportions.iloc[:, 1:].div(proportions.iloc[:, 1:].sum(axis=1), axis=0)
```

```python
proportions= proportions.rename(columns={"batter":"player_id"})
```

```python
df_2025 = df_2025.merge(proportions, how='inner', on='player_id')
```

```python
df_2025['pull_air']=df_2025['Pull_fly_ball']+df_2025['Pull_line_drive']
df_2025['oppo_air']=df_2025['Oppo_fly_ball']+df_2025['Oppo_line_drive']
df_2025['center_air']=df_2025['Center_fly_ball']+df_2025['Center_line_drive']
```

```python
print(ss.pearsonr(x=df_2025['est_woba_minus_woba_diff'], y=df_2025['pull_air']))
print(ss.pearsonr(x=df_2025['est_woba_minus_woba_diff'], y=df_2025['center_air']))
ss.pearsonr(x=df_2025['est_woba_minus_woba_diff'], y=df_2025['oppo_air'])
```
