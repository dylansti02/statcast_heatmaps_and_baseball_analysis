```python
import pybaseball as pb
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as ss
import matplotlib.ticker
```

```python
pitchers_25 = pb.pitching_stats(2025, qual = 20)
```

```python
pitchers_velo = pb.statcast_pitcher_pitch_arsenal(2025)
```

```python
pitchers_25['ff_avg_speed']=pitchers_velo['ff_avg_speed']
```

```python
sns.scatterplot(
    data=pitchers_25,
    x='ff_avg_speed',
    y='BB%'
)
```
